```julia
# Shift all pixels of i1 to the right by the value of gt
function shift_disparity(i1::Array{Float64,2}, gt::Array{Float64,2})
    
    @assert size(i1) == size(gt)
    
    
    id = zeros(size(i1))
    
    for x in 1:size(i1,2)
        for y in 1:size(i1,1)            
            if gt[y,x]!=0
                #id[y,x]=0
                id[y,x+convert(Int32,gt[y,x])]= i1[y,x]
                
            end

        end
    end
            
        
    @assert size(id) == size(i1)
    
    return id::Array{Float64,2}
end
```

```julia
function r2G(I::Array{Float32,3})
    I=convert(Array{Float64,3}, I)
    I_gray = 0.2989*I[:,:,1] + 0.5870*I[:,:,2] + 0.1140*I[:,:,3]
    return I_gray
end
```

```julia
dm=PyPlot.imread("gt.png")
dm=dm*255
dm=convert(Array{Float64,2}, dm)

p1=PyPlot.imread("i0.png")
p1=r2G(p1)
```

```julia
PyPlot.imshow(p1,"gray")
```

```julia
PyPlot.imshow(dm)
```

```julia
shifted=shift_disparity(p1,dm)
PyPlot.imshow(shifted,"gray")
```
